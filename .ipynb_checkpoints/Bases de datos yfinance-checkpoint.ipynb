En la explicación que muestras, los conceptos clave son:

🔹 Módulo
Un módulo en Python es un archivo que contiene código Python, como funciones, clases y variables, que puedes importar y usar en tu propio código.

📌 Ejemplo en tu caso:
yfinance es un módulo que proporciona herramientas para obtener datos financieros.

🔹 Clase
Una clase es una plantilla para crear objetos en Python. Define atributos (datos) y métodos (funciones) que los objetos creados a partir de la clase pueden usar.

📌 Ejemplo en tu caso:
Ticker es una clase dentro del módulo yfinance. Define propiedades y métodos para trabajar con información sobre un activo financiero.

vist = yf.Ticker('VIST')
Aquí, Ticker es la clase que se usa para crear el objeto vist.

```python
metodos = [metodo for metodo in dir(vist) if not metodo.startswith("_")]
for metodo in metodos:
    atributo = getattr(vist, metodo)
    print(f'{metodo}: {type(atributo)}')
```

```python
help(vist.history)
```

```python
targets = vist.analyst_price_targets
print(targets.keys())  # Muestra las claves disponibles

# Acceder a datos específicos (si existen)
print(targets.get("current"))
print(targets.get("high"))  # Precio objetivo alto
print(targets.get("low"))   # Precio objetivo bajo
print(targets.get("mean"))  # Precio objetivo promedio
print(targets.get("median")) # Mediana de los precios objetivo
```
