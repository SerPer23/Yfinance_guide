```python
# Función para intentar obtener datos de cada método y manejar excepciones
def print_method_data(method, title):
    try:
        print(f"\n{title}:")
        data = getattr(ticker, method)()  # Usamos getattr para llamar el método dinámicamente
        print(data)
    except Exception as e:
        print(f"\n{title}: No disponible o no se pudo obtener datos. Error: {e}")

# Función para manejar atributos no métodos (como `isin` o `proxy`)
def print_attribute_data(attribute, title):
    try:
        print(f"\n{title}:")
        data = getattr(ticker, attribute)  # Usamos getattr para acceder al atributo
        print(data)
    except Exception as e:
        print(f"\n{title}: No disponible o no se pudo obtener datos. Error: {e}")

# Llamada a los métodos y atributos disponibles

print_method_data("get_insider_purchases", "Insider Purchases")
print_method_data("get_insider_roster_holders", "Insider Roster Holders")
print_method_data("get_insider_transactions", "Insider Transactions")
print_method_data("get_institutional_holders", "Institutional Holders")
print_attribute_data("isin", "ISIN")
print_method_data("get_major_holders", "Major Holders")
print_method_data("get_mutualfund_holders", "Mutual Fund Holders")
print_method_data("get_news", "News")
print_method_data("option_chain", "Option Chain")
print_attribute_data("options", "Options")
print_attribute_data("proxy", "Proxy")
```
